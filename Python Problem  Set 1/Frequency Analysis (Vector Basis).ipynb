```python
import nltk
nltk.download('punkt')
```

```python
from nltk import tokenize
from sklearn.feature_extraction.text import CountVectorizer
from scipy import spatial
import random
vectorizer = CountVectorizer()
```

```python
def obtainTextData(novel):
    sentences = tokenize.sent_tokenize(novel)
    words = novel.split()
    return [sentences, words]
```

```python
def cosineSimCheck(novel, sentence):
  novel += sentence
  sentences = tokenize.sent_tokenize(novel)
  senVectors = vectorizer.fit_transform(sentences)
  senVectors = senVectors.toarray()
  sentence = senVectors[len(senVectors)-1]
  cosSimilarity = 0
  count = 0
  while cosSimilarity < .99 or count == 20:
    randomSentence = senVectors[random.randint(0, len(senVectors) - 1)]
    #Had to look this up
    cosSimilarity = 1 - spatial.distance.cosine(sentence, randomSentence)
    count += 1
  return cosSimilarity >= 1
```

```python
def makeRandomSentence(words, sentenceVectors):
  sentenceLength = len(words)/len(sentenceVectors)
  sentenceLength = random.randint(10, 15)
  sentence = ""
  for i in range(sentenceLength):
    randomWord = random.randint(0, len(words)-1)
    sentence += " " + words[randomWord]

  return sentence +"."
```

```python
def obtainTextGeneration(words, sentenceVectors, book, sentencesDesired):
  textGeneration = []
  for i in range(sentencesDesired):
    sentence = makeRandomSentence(words, sentenceVectors)
    sentenceGood = cosineSimCheck(book, sentence)
    while (sentenceGood == False):
      sentence = makeRandomSentence(words, sentenceVectors)
      sentenceGood = cosineSimCheck(book,sentence)
    textGeneration.append(sentence)
  finalText = ""
  for i in range(len(textGeneration)):
    finalText += " " + textGeneration[i] +"\n"
  return finalText
```

```python
def textPreprocess(bookString):
  bookString = bookString.lower()
  bookString = bookString.replace("(", "")
  bookString = bookString.replace(")", "")
  bookString = bookString.replace(",", "")
  return bookString
```

```python
filepath = "Twilight.txt"
book = open(filepath, 'r')
book = book.read()
```

```python
bookString = textPreprocess(book)
```

```python
sentenceVectors, words = obtainTextData(bookString)
textGeneration = obtainTextGeneration(words, sentenceVectors, bookString, 50)
```
