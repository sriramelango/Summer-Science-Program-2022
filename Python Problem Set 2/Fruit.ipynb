```python
# Slicing and Dicing
# Fruit
# 6/27/22
# Sriram Elango

import numpy as np

fruits = np.array([["Apple","Banana","Blueberry","Cherry"],
["Coconut","Grapefruit","Kumquat","Mango"],
["Nectarine","Orange","Tangerine","Pomegranate"],
["Lemon","Raspberry","Strawberry","Tomato"]])
```

```python
print("Section A: " + fruits[len(fruits) -1][len(fruits[0]) -1])
```

```python
print("Section B: ")
sectionB = fruits[1:3, 1:3]
print(sectionB)
```

```python
print("Section C: ")
sectionC = fruits[0:3:2]
print(sectionC)
```

```python
print("Section D: ")
sectionD = fruits[1:3, 1:3][::-1, ::-1]
print(sectionD)
```

```python
print("Section E: ")
sectionEFinal = fruits.copy()
sectionEFinal[:, [0, 3]] = sectionEFinal[:, [3,0]]
print(sectionEFinal)
```

```python
print("Section F: ")
fruits[0:4, 0:4] = "Sliced!"
print(fruits)
```
