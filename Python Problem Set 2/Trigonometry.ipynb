```python
# Calc two angles and a side
# Sriram Elango
# 6/27/2022

from math import asin, acos, pi, cos, sin

# a function to determine the quadrant of an angle based on its sine and cosine (in radians)
# returns the angle in the correct quadrant (in radians)
def findQuadrant(sine, cosine):
    if cosine > 0 and sine > 0: #1
        return asin(sine)

    if cosine < 0 and sine > 0: #2
        return acos(cosine)

    if cosine < 0 and sine < 0: #3
        return pi - asin(sine)

    if cosine > 0 and sine < 0: #4
        return 2*pi + asin(sine)
```

```python
# a function that given the values (in decimal degrees) of two sides and the included angle of a spheical triangle
# returns the values of the remaining side and two angles (in decimal degrees)

def radiansToDegrees(radians):
  return (radians * 180)/pi

def degreesToRadians(degrees):
  return (degrees * pi)/180

def SAS(a, B, c):
  a = degreesToRadians(a)
  B = degreesToRadians(B)
  c = degreesToRadians(c)

  b = acos(((cos(c) * cos(a)) + (sin(c) * sin(a) * cos(B))))
  sinA = (sin(B) * sin(a))/sin(b)
  sinC = (sinA)*sin(c)/sin(a)

  cosA =  (cos(a)-cos(b)*cos(c))/(sin(b) * sin(c))
  cosC = (cos(c) - cos(a)*cos(b)/(sin(a) * sin(b)))

  A = findQuadrant(sinA, cosA)
  C = findQuadrant(sinC, cosC)



  b = radiansToDegrees(b)
  C = radiansToDegrees(C)
  A = radiansToDegrees(A)        
 
  return A,b,C
```
