```python
import numpy as np
import math
```

```python
taus = [-0.10432057611305061, 0.2727289257369039, 0.3770495018499545]
Sun2 = [-0.133229582916348, 0.9247467444400224, 0.4008445345253298]
rhohat2 = [-0.14399356895854978, -0.9289800184607053,
-0.34097210648281323]
Ds = [-0.0015053745546385585, -0.033894368435358625,
-0.05456332153042902, -0.10536229171721766]
```

```python
def obtainMagnitude(vector):
  vectorMag = math.sqrt(vector[0]**2 + vector[1]**2 + vector[2]**2)
  return vectorMag
```

```python
def SEL(taus, Sun2, rhohat, Dvalues):
  taus = np.array(taus)
  Sun2 = np.array(Sun2)
  rhohat = np.array(rhohat)

  mu = 1
  t = taus[2]
  a1 = taus[1]/t 
  b1 = (a1/6)*(t**2 - taus[1]**2)
  a3 = (-taus[0])/t
  b3 = ((a3)/6) * (t**2 - taus[0]**2)

  A = (a1 * Dvalues[1] - Dvalues[2] + a3 * Dvalues[3])/(- Dvalues[0])
  B = (b1 * Dvalues[1] + b3 * Dvalues[3])/( - Dvalues[0])

  F = obtainMagnitude(Sun2)**2
  E = -2 * (np.dot(rhohat,Sun2))
  
  aCoeff = - (A**2 + A * E + F)
  bCoeff = - mu * (2 * A * B + B * E)
  cCoeff = - (mu ** 2) * (B ** 2)
  coEffArray = np.asarray([1, 0, aCoeff, 0, 0, bCoeff, 0, 0, cCoeff])

  print(coEffArray)

  roots = np.roots(coEffArray)
  print(roots)

  #Obtain only real positive numbers
  for i in range(len(roots)):
    if ((np.isreal(roots[i]) == False)):
      roots[i] = 0

  roots = roots[roots != 0]
  roots = roots[roots >= 0]
  roots = roots.astype(float)

  #Obtain Rhos

  rhos = []
  for i in range(len(roots)):
    rhos.append(A + (B/((roots[i])**3)))
  
  rhos = np.array(rhos)

  #Filter Roots from negative Rhos / negative ranges

  for i in range(len(rhos)):
    if (rhos[i] < 0):
      roots[i] = 0

  roots = roots[roots != 0]
  rhos = rhos[rhos >= 0]

  return roots, rhos
```
